```python
num = int(input("Enter number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
n = int(input("Enter n: "))
total = 0

for i in range(1, n + 1):
    total += i

print("Sum =", total)
```

```python
n = int(input("Enter rows: "))

for i in range(1, n + 1):
    for j in range(i):
        print("*", end=" ")
    print()
```
