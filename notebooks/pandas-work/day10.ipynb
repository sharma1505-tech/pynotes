```python
text = input("Enter text: ").lower()

vowels = "aeiou"
v_count = 0
c_count = 0

for ch in text:
    if ch.isalpha():
        if ch in vowels:
            v_count += 1
        else:
            c_count += 1

print("Vowels:", v_count)
print("Consonants:", c_count)
```

```python
text = input("Enter text: ")

rev = ""
for ch in text:
    rev = ch + rev

print("Reversed:", rev)
```

```python
text = input("Enter text: ")

rev = text[::-1]

if text == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```
