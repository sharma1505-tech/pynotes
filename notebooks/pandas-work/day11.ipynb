```python
def calculator(a, b):
    print("Addition:", a + b)
    print("Subtraction:", a - b)
    print("Multiplication:", a * b)
    
    if b != 0:
        print("Division:", a / b)
    else:
        print("Division not possible")

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

calculator(x, y)
```

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

num = int(input("Enter number: "))
print("Factorial:", factorial(num))
```

```python
def is_prime(n):
    if n <= 1:
        return False
    
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = int(input("Enter number: "))

if is_prime(num):
    print("Prime Number")
else:
    print("Not Prime")
```
