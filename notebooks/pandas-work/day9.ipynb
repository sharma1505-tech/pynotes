```python
nums = list(map(int, input("Enter numbers: ").split()))
unique = []

for num in nums:
    if num not in unique:
        unique.append(num)

print("List without duplicates:", unique)
```

```python
nums = list(map(int, input("Enter numbers: ").split()))

largest = second = -999999

for num in nums:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num

print("Second largest:", second)
```
