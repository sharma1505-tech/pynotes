```python
while True:
    msg = input("You: ")
    
    if msg == "hi":
        print("Bot: Hello!")
    elif msg == "how are you":
        print("Bot: I'm fine!")
    elif msg == "bye":
        print("Bot: Bye!")
        break
    else:
        print("Bot: I don't understand")
```

```python
msg = input("Enter message: ")

if "python" in msg:
    print("Python is a programming language")
else:
    print("Unknown topic")
```

```python
mood = input("How are you feeling? ")

if mood == "happy":
    print("Great! Keep smiling 😊")
elif mood == "sad":
    print("Don't worry, things will be fine ❤️")
else:
    print("Stay positive!")
```
