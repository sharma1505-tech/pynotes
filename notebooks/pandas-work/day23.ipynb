```python
n = int(input("Enter rows: "))

for i in range(1, n + 1):
    print(" " * (n - i), end="")
    
    for j in range(1, i + 1):
        print("*", end=" ")
    print()
```

```python
n = int(input("Enter rows: "))
num = 1

for i in range(1, n + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
n = int(input("Enter rows: "))

for i in range(n):
    num = 1
    for j in range(i + 1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)
    print()
```
