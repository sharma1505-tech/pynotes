```python
n = int(input("Enter number of elements: "))
nums = []

for i in range(n):
    val = int(input("Enter number: "))
    nums.append(val)

print("List:", nums)
```

```python
n = int(input("Enter rows: "))

for i in range(n, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
n = int(input("Enter rows: "))

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```
