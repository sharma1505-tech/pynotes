```python
student = {}

name = input("Enter name: ")
marks = int(input("Enter marks: "))

student["name"] = name
student["marks"] = marks

print("Student Details:", student)
```

```python
text = input("Enter sentence: ")
words = text.split()

count = {}

for word in words:
    if word in count:
        count[word] += 1
    else:
        count[word] = 1

print("Word Count:", count)
```

```python
data = {"a": 10, "b": 20, "c": 30}

key = input("Enter key to search: ")

if key in data:
    print("Value:", data[key])
else:
    print("Key not found")
```
